```python
#1.9构建线性回归
#1.10预测输出
#1.11模型评估
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.max_colwidth',None)
data = pd.read_csv('bike.csv') 
data=data.drop('id',axis=1)
data=data[data['city']==1]
data=data.drop('city',axis=1)
def replace(x):
    if x>=6 and x<=18:
        return 1
    else:
        return 0
data['hour']=data['hour'].apply(replace)
y_vector=data['y'].to_numpy().reshape(-1,1)
data=data.drop('y',axis=1)
data_array=data.to_numpy()
train_data,test_data=train_test_split(data,test_size=0.2,random_state=20)
train_vector,test_vector=train_test_split(y_vector,test_size=0.2,random_state=20)
scaler1=MinMaxScaler()
scaler2=MinMaxScaler()
scaler3=MinMaxScaler()
scaler4=MinMaxScaler()
train_data=pd.DataFrame(scaler1.fit_transform(train_data))
train_vector=pd.DataFrame(scaler2.fit_transform(train_vector))
test_data=pd.DataFrame(scaler3.fit_transform(test_data))
test_vector=pd.DataFrame(scaler4.fit_transform(test_vector))

model=LinearRegression()
model.fit(train_data,train_vector)

y_pred=model.predict(test_data)
print("结果预测：")
print(y_pred)
print()

mse=mean_squared_error(test_vector,y_pred)
print("均方根误差（RMSE）：")
print(np.sqrt(mse))
```

```python
#2.3主成分分析
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# 加载鸢尾花数据集
iris = load_iris()
data = iris.data
target = iris.target
# 数据标准化
scaler=StandardScaler()
X_scaled=scaler.fit_transform(data)
# 主成分分析
pca = PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)

print("Variance ratio of each component:", pca.explained_variance_ratio_)
print("Cumulative variance ratio:", np.cumsum(pca.explained_variance_ratio_))

for i in range(len(np.unique(target))):
    plt.scatter(X_pca[target == i, 0], X_pca[target == i, 1], 
                label=iris.target_names[i], alpha=0.7)

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Iris Dataset')
plt.legend()
plt.grid()
plt.show()
```

```python
#2.4以k=3训练
#2.5输出准确率
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 加载鸢尾花数据集
iris = load_iris()
data = iris.data
target = iris.target
# 拆分数据集为训练集和测试集
train_data, test_data, train_target, test_target = train_test_split(data, target, test_size=0.8, random_state=42)

# 训练KNN模型
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(train_data, train_target)

# 使用模型进行预测
y_pred = knn.predict(test_data)

# 评估模型
accuracy = accuracy_score(test_target, y_pred)
print(f"准确率： {accuracy:.2f}")
```
