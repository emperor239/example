```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=[13,8]

# 读取数据
df=pd.read_csv('users_combined_info_500.csv')

# 生成图像
x=df['country'].value_counts()
x.plot(kind='bar',rot=90,legend=True,
       title="Different countries",
       xlabel="label",ylabel="number")

# 设置图表标题和标签
plt.title('Distribution of different countrys')
plt.xlabel('Country')
plt.ylabel('Count')

# 显示图表
print('1.1')
print('用户所在国家和地区的分布：')
plt.show()
print('以下为主要的开发者集中地，与其提交数：')
print(x.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=[13,8]

# 读取数据
df=pd.read_csv('users_combined_info_500.csv')
df=pd.to_datetime(df['event_time'])

# 提取时区信息并存储在新的列中
df = df.apply(lambda x: str(x.hour))

x=df.value_counts()
x.plot(kind='bar',rot=0,legend=True,
       title="Different countries",
       xlabel="label",ylabel="number")

# 设置图表标题和标签
plt.title('Distribution of different event_time hours')
plt.xlabel('Event_time hour')
plt.ylabel('Count')

# 显示图表
print('1.3')
print('用户的时区分布：')
print('基于东八区区时下的，各地区协作时间模式：')
plt.show()
print('可以看到，假设人们在白天工作，基于中国时间，那么23~次日6点人数是最少的，其他时间人数较多，这表明欧亚地区区时所对应的所有国家的协作人数总数较大，欧美大陆区时所对应的国家的单国家协作人数较多')

```
